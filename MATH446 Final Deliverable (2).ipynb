# Introduction

## Data Dictionary
- `action_name`: BURN, MINT, or SWAP <br>
- `timestamp`: Unix time <br>
- `newly_added_usdc`: positive/negative number according to burn/mint/swap activities <br>
- `newly_added_eth`: positive/negative number according to burn/mint/swap activities <br>
- `transaction.id`: wallet ID for each wallet in the pool <br>
- `cum_sum_ETH`: previous culmulative ETH + ETH added <br>
- `cum_sum_USDC`: previous culmulative USDC + USDC added <br>
- `ETH Rate`: use `cum_sum_USDC` / `cum_sum_ETH` to get how much dollar 1 ETH worth with respect to different liquidity activities <br>
- `personal_perc`: front_fill() for swap (ownership does not change for swaps), burn and mint activities with functions below <br>
- `impermanent_loss`: 1 - liquidity providing value / hold value <br>
- `amount0In`, `amount1In`, `amount0Out`, `amount1Out`: coin 0 (usdc) and coin 1 (ETH) in and out for SWAP action <br>
- `amount0`, `amount1`: coin 0 (usdc) and coin 1 (ETH) in and out for BURN and MINT action, burn is negative and mint is positive <br>
- `transaction_fee_gained`: transaction fee gained after swaps; note: there's a 0.3% fee for swaps, 0.25% of fee given to LP, remaining 0.05% to SUSHI token holders <br>

## Additional Notes 
1. ETH to USDC ratio should not change for the burn and mint activities <br>
2. For simplicity, we assume each person enters the pool once and stays in the pool <br>
3. USDC price will be calculated as 1 dollar <br>
4. Sushi Swap Data is initiated with a SWAP, therefore an temporary initial liquidity is calculated as below: <br>

assume x = initial USDC and y = initial ETH <br>
we can come up with a system of equations from the equality: `(x + swaps_x) / (y + swaps_y) = (next_mint_x) / (next_mint_y)` <br>
using the first two sets of swaps and burn/mint, we get a result: approximately 410724337.7 USDC and 1112980.847 ETH at the beginning of the pool

## Simplify actions dataset
Next, we created a dataset which combines burn, mint, and swap transactions into a simple +/- change of DAI and ETH in the pool. Start by filling in NaN's with 0s in order to add and subtract the total change in a transaction. We can do this because a burn or mint will have 0 in swap columns and vice versa. 

```python
dai_eth_actions['LOGGED_AMOUNT0'] = dai_eth_actions['LOGGED_AMOUNT0'].fillna(0).astype(float)
dai_eth_actions['LOGGED_AMOUNT1'] = dai_eth_actions['LOGGED_AMOUNT1'].fillna(0).astype(float)
dai_eth_actions['AMOUNT0_IN'] = dai_eth_actions['AMOUNT0_IN'].fillna(0).astype(float)
dai_eth_actions['AMOUNT1_IN'] = dai_eth_actions['AMOUNT1_IN'].fillna(0).astype(float)
dai_eth_actions['AMOUNT0_OUT'] = dai_eth_actions['AMOUNT0_OUT'].fillna(0).astype(float)
dai_eth_actions['AMOUNT1_OUT'] = dai_eth_actions['AMOUNT1_OUT'].fillna(0).astype(float)
```

For each token, we perform the following: burn/mintDAI + swapDAIin - swapDAIout. Simply put, this gives us the amount of DAI and ETH involved in the transaction, including a negative number for the amount that goes out of the pool in a swap. We will go back and switch the burns to a negative in the next step. We divide by 10^18 to get it into the actual number form.

```python
dai_eth_actions['DAI_CHANGE'] = (dai_eth_actions['LOGGED_AMOUNT0']+dai_eth_actions['AMOUNT0_IN']-dai_eth_actions['AMOUNT0_OUT'])/(10**18)
dai_eth_actions['ETH_CHANGE'] = (dai_eth_actions['LOGGED_AMOUNT1']+dai_eth_actions['AMOUNT1_IN']-dai_eth_actions['AMOUNT1_OUT'])/(10**18)
```

Use a for loop to identify all burn transactions and change the sign to a negative number, reflecting the shrinking of the pool when burns take place. 

```python
for i in range(1,len(dai_eth_actions)):
    if dai_eth_actions.loc[i]['EVENT_TYPE'] == 'BURN':
        dai_eth_actions.at[i,'DAI_CHANGE'] = -1* dai_eth_actions.at[i,'DAI_CHANGE']
        dai_eth_actions.at[i,'ETH_CHANGE'] = -1* dai_eth_actions.at[i,'ETH_CHANGE']
```

Create simplified dataframe to show changes in ETH and DAI in pool for each given transaction

```python
change_in_pool = dai_eth_actions.drop(['AMOUNT0_IN', 'AMOUNT1_IN', 'AMOUNT0_OUT', 'AMOUNT1_OUT','TOKEN_1','TOKEN_0','LOGGED_AMOUNT0','LOGGED_AMOUNT1','POOL_DECIMALS','POOL_NAME','TX_HASH','TOKEN_0_TICKER_SYMBOL','TOKEN_1_TICKER_SYMBOL','ADDRESS_FROM','ADDRESS_TO'],axis=1)
change_in_pool
```

## Initialize Impermanent loss functions

### Calculation for `personal_perc`

When a new id mint certain percentage of coins into the pool, the person only owns previous_percentage * (1 - new_mint_perc) of the pool

```python
def mint_percentage_change(new_eth, new_total_eth, previous_personal_percentage):
    mint_percentage = new_eth / new_total_eth
    return previous_personal_percentage * (1 - mint_percentage)
```

When a new id burn certain percentage of coins into the pool, the person will own previous_percentage * (1 + new_mint_perc) of the pool

```python
def burn_percentage_change(new_eth, new_total_eth, previous_personal_percentage):
    burn_percentage = new_eth / new_total_eth
    return previous_personal_percentage * (1 + burn_percentage)
```

Note: swaps won't result in the change of personal percentage of the pool because the AMM curve does not move

```python
# Calculates personal percentage using subsets of data from from all_data
def calculate_personal_percentage(df):
    i = 1
    while i < len(df):
        if df['action_name'][i] == "MINT":
            mint_perc_change = mint_percentage_change(new_eth = abs(df['amount1'][i]), 
                                   new_total_eth = df['cum_sum_eth'][i], 
                                   previous_personal_percentage = df['personal_perc'][i-1])
            df['personal_perc'][i] = mint_perc_change
        elif df['action_name'][i] == 'BURN':
            burn_perc_change = burn_percentage_change(new_eth = abs(df['amount1'][i]), 
                                   new_total_eth = df['cum_sum_eth'][i], 
                                   previous_personal_percentage = all_data['personal_perc'][i-1])
            df['personal_perc'][i] = burn_perc_change
        elif df['action_name'][i] == "SWAP":
            df['personal_perc'][i] = df['personal_perc'][i-1]
        i+=1
        # print(df['personal_perc'][i])
    return df
```

### Calculation for `impermanent_loss`

To calculate `impermanent_loss`, filter swaps rows and plug in all the corresponding parameters to the function

```python
# personal ownership column, refer to the person_eth and person_usdc at the very beginning
def hold_value(person_eth, person_usdc, new_usdc, new_eth, usdc_price):
    usdc_price = 1
    eth_rate = 1792
    return person_eth * eth_rate * usdc_price + person_usdc * usdc_price # eth_rate is the new rate using new usdc / eth;

def liquidity_value(person_percentage, new_usdc, new_eth, usdc_price):
    eth_rate = new_usdc / new_eth
    return (person_percentage/100) * new_eth * eth_rate * usdc_price + (person_percentage/100) * new_usdc * usdc_price


# if swap, impermanent loss will change, personal percentage won't change
# start_eth_rate = last swap usdc/eth; end_eth_rate = curr swap usdc/eth
def impermanent_loss_calculation(person_eth, person_usdc, new_usdc, new_eth, usdc_price, person_percentage, eth_added):
    # exchange rate 1 ETH = how much USDC
    hold = hold_value(person_eth, person_usdc, new_usdc, new_eth, usdc_price)
    lp = liquidity_value(person_percentage, new_usdc, new_eth, usdc_price)
    impermanentLoss = 1 - lp / hold
    transaction_fee_gained = person_percentage * abs(eth_added) * 0.003 # transaction fee # or * 0.0025 for sushiswaps
    return impermanentLoss, transaction_fee_gained
```

```python
# Adds columns with impermanent loss and transaction fee with a subset of all_data
# Inputs: 
    # df = DataFrame taking the same schema as all_data, filtered to all swaps occuring after the first mint for a given Wallet ID
    # eth_value = initial ETH value for that wallet address
    # usdc_value = initial USDC value for that wallet address
def calculate_impermanent_loss(df, eth_value, usdc_value):
    df['impermanent_loss'] = 0.0
    df['transaction_fee_gained'] = 0.0
    i = 0
    while i < len(df):
        impermanentLoss, transaction_fee_gained = impermanent_loss_calculation(person_eth = eth_value,
                                                                      person_usdc = usdc_value,
                                                                      new_usdc = df['cum_sum_usdc'][i],
                                                                      new_eth = df['cum_sum_eth'][i], 
                                                                      usdc_price = 1,
                                                                      person_percentage = df['personal_perc'][i],
                                                                      eth_added = max(abs(df['amount1'][i]), abs(df['amount1In'][i]), abs(df['amount1Out'][i])))
        df['impermanent_loss'][i] = impermanentLoss
        df['transaction_fee_gained'][i] = transaction_fee_gained
        i+=1
    return df
```

## Gather all data for wallets of interest

```python
mints_of_interest = all_data.loc[all_data['to'].isin(wallets_of_interest)]
burns_of_interest = all_data.loc[all_data['sender'].isin(wallets_of_interest)]
frames_int = [burns_of_interest,mints_of_interest] # this actually unintentionally gathered all swaps too for these wallets
new_data = pd.concat(frames_int) # use same code from above to combine into one dataframe
new_data = new_data.sort_index() # sort so we have chronological order instead of mints then burns
new_data = new_data.reset_index() #gets rid of old index
```

```python
burn_mints = new_data[new_data['action_name'] != 'SWAP'] # get rid of swaps
burn_mints = burn_mints.drop(['level_0', 'liquidity', 'index', 'complete', 'feeLiquidity', 'feeTo', 'id', 'logIndex', 'transaction.id', 'cum_sum_usdc', 'cum_sum_eth', 'personal_perc'], axis=1)
burn_mints = burn_mints.reset_index(drop=True)
burn_mints.head()
```

We used this for loop to gather the data from new_data (which is grouped by transaction) into our dataframe that has information about the wallets. By generating lists for all the relevant data, and then appending it to the wallets_of_interest dataframe, we are able to have all relevant data in one place, based on wallet address. 

```python
# calculate the time in pool & actual gains
time_in_pool = []
actual_gains = []
burn_times = []
mint_times = []
mint_amount_eth = []
mint_amount_usdc = []
burn_amount_eth = []
burn_amount_usdc = []
for wallet in wallets_of_interest:
    for i in range(len(burn_mints)):
        if burn_mints['action_name'][i] != 'SWAP':
            if burn_mints['sender'][i] == wallet:
                burn_time = burn_mints['timestamp'][i]
                amount_out = burn_mints['amountUSD'][i]
                burn_times.append(burn_time)  
                eth_burned = - burn_mints['newly_added_eth'][i]
                burn_amount_eth.append(eth_burned)
                usdc_burned = - burn_mints['newly_added_usdc'][i]
                burn_amount_usdc.append(usdc_burned)
            elif burn_mints['to'][i] == wallet:
                mint_time = burn_mints['timestamp'][i]
                amount_in = burn_mints['amountUSD'][i]
                mint_times.append(mint_time)
                eth_minted = burn_mints['newly_added_eth'][i]
                mint_amount_eth.append(eth_minted)
                usdc_minted = burn_mints['newly_added_usdc'][i]
                mint_amount_usdc.append(usdc_minted)
    time_stayed = burn_time - mint_time
    time_in_pool.append(time_stayed)   
    actual_gain = amount_out - amount_in
    actual_gains.append(actual_gain)
```

```python
oneburnonemint['time_in_pool'] = time_in_pool
oneburnonemint['actual_gain'] = actual_gains
oneburnonemint['mint_time'] = mint_times
oneburnonemint['burn_time'] = burn_times
oneburnonemint['eth_minted'] = mint_amount_eth
oneburnonemint['usdc_minted'] = mint_amount_usdc
oneburnonemint['eth_burned'] = burn_amount_eth
oneburnonemint['usdc_burned'] = burn_amount_usdc
```
